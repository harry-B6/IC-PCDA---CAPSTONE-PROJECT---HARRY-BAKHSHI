Create function for formatting Plotly plots:

```python
#For formatting plots:
def format_plot(fig, legends = None, xlabel="Time", ylabel="Value", figsize=(500,900), font_size=15, title_font_size=20):
    if legends:
        names = cycle(legends)
        fig.for_each_trace(lambda t:  t.update(name = next(names)))
    fig.update_layout(
            autosize=False,
            width=figsize[1],
            height=figsize[0],
            title={
            'x':0.5,
            'xanchor': 'center',
            'yanchor': 'top'},
            titlefont={
                "size": 20
            },
            legend_title = None,
            legend=dict(
                font=dict(size=font_size),
                orientation="h",
                yanchor="bottom",
                y=0.98,
                xanchor="right",
                x=1,
            ),
            yaxis=dict(
                title_text=ylabel,
                titlefont=dict(size=font_size),
                tickfont=dict(size=font_size),
            ),
            xaxis=dict(
                title_text=xlabel,
                titlefont=dict(size=font_size),
                tickfont=dict(size=font_size),
            )
        )
    return fig

#^https://github.com/PacktPublishing/Modern-Time-Series-Forecasting-with-Python/blob/main/notebooks/Chapter03/01-Visualizing%20Time%20Series.ipynb
#Accessed 06/06/2024
```

```python
df = metric_df[metric_df['Lasso Regression'] == 1].copy()
drop_list = []
for col in range(len(list(df.columns.values)[:5])):
    ser = df[list(df.columns.values)[col]]
    for row in range(len(df)):
        if pd.isnull(ser.iloc[row]):
            drop_list.append(str(col) + ', ' + str(row)) 
found_list = []
unique_list = []
entry_list = drop_list
entry = []
entry_list_length = len(entry_list)
#Find unique members of the list:
for i in range(entry_list_length):
    entry = entry_list[i]
    if entry not in found_list:
        unique_list.append(entry)
        found_list.append(entry)
drop_list = unique_list
for _null in drop_list:
    if int(_null[3]) in df.index:
        df.drop(int(_null[3]), inplace=True)
    else:
        pass
df.set_index(np.arange(0, len(df)), inplace=True)
df
found_list = []
unique_list = []
hyp = 'max iter'
entry_list = list(df[hyp])
entry = []
entry_list_length = len(entry_list)
#Find unique members of the list:
for i in range(entry_list_length):
    entry = entry_list[i]
    if entry not in found_list:
        unique_list.append(entry)
        found_list.append(entry)
df['median_unique_hyp_MAE'] = np.nan
df['median_unique_hyp_MSE'] = np.nan
df['median_unique_hyp_MASE'] = np.nan
df['median_unique_hyp_Forecast_Bias'] = np.nan
for unique_hyp in range(len(unique_list)):
    for row in range(len(df)):
        if df[hyp][row] == unique_list[unique_hyp]:        
            df['median_unique_hyp_MAE'][row] = statistics.median(df[df[hyp] == unique_list[unique_hyp]]['MAE'])
            df['median_unique_hyp_MSE'][row] = statistics.median(df[df[hyp] == unique_list[unique_hyp]]['MSE'])
            df['median_unique_hyp_MASE'][row] = statistics.median(df[df[hyp] == unique_list[unique_hyp]]['MASE'])
            df['median_unique_hyp_Forecast_Bias'][row] = statistics.median(df[df[hyp] == unique_list[unique_hyp]]['Forecast Bias'])
file = '/Users/harrybakhshi/Desktop/Python_notes/data/PCDA_capstone/hyperparameter_tuning/test_pred_lasso_max_iter_metric_df.pik'
with open(file, 'wb') as f:
    pickle.dump(df, f) #write df to .pik file on disk
# with open(file, 'rb') as f:
#       df = pickle.load(f) #load pickle file 'file' into variable
```
